```python
# 기본 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프 기본 테마 설정
sns.set()

# 그래프 기본 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

# 복잡한 통계 처리를 위한 라이브러리
from scipy import stats
```

```python
# data 불러오기
df = pd.read_csv('data/2324.csv')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)

#print(df.isnull().sum())
df.tail(2)
```

```python
df2324 = df[df['act_interest'].notna()]

print(df2324.shape)
print(df2324['act_interest'].isna().sum())

df2324 = df[df['act_interest'].notna()]
print(df2324['act_interest'].isna().sum())  # 0이어야 정상
print(df['act_interest'].dtype)

df2324.isna().sum()  # 결측치 확인
```

```python
# 웰니스 변수 만들기
df2324['wellness'] = ((df2324['act_interest'] == 3) | (df2324['act_interest'] == 12)).astype(int)
df2324['wellness'].value_counts() # 웰니스 관광객 4160명
```

```python
# 정보부족 여부
print(df2324['info_1'].value_counts())
print(df2324['info_2'].value_counts())
print(df2324['info_3'].value_counts())

print(df2324['info_1'].isna().sum())
print(df2324['info_2'].isna().sum())
print(df2324['info_3'].isna().sum())
```

```python
# 정보 부족 항목 수 계산: info_lack_1~3 중 Not-NaN의 개수 세기
df2324['infolack'] = df2324[['info_lack_1', 'info_lack_2', 'info_lack_3']].notna().sum(axis=1)
print(df2324['infolack'].value_counts())

# 0: 아무 항목도 선택하지 않음 (무응답, 이 경우에 어떻게 해석해야 할지)
# 1: 3개 중 1개 선택 -> 정보 부족 경험 있음
# 2 : 3개 중 2개 선택 -> 더 많은 정보 부족
# 3 : 3개 모두 선택 -> 정보 부족 경험 강함

```

### 0. 주요 변수
1. overall_satis: 1	매우 불만족, 2 불만족, 3 보통, 4 만족, 5 매우 만족
2. infolack: 0(무응답), 1(1/3), 2(2/2), 3(3/3, 정보 부족에 갈망을 느끼는 집단)
3. wellness : 0(일반), 1(웰니스)

### 1. 통계검정

```python
from scipy.stats import shapiro, normaltest, anderson, levene
```

```python
group_no_lack = df2324[df2324['infolack'] == 0]['overall_satis'].dropna()
group_lack1 = df2324[df2324['infolack'] == 1]['overall_satis'].dropna()
group_lack2 = df2324[df2324['infolack'] == 2]['overall_satis'].dropna()
group_lack3 = df2324[df2324['infolack'] == 3]['overall_satis'].dropna()

# 정규성 검정
for i, group in zip([0,1,2,3], [group_no_lack, group_lack1, group_lack2, group_lack3]):
    stat, p = shapiro(group)
    print(f'infolack={i} 정규성 검정: W={stat:.3f}, p={p:.4f}')

# 등분산성 검정
stat, p = levene(group_no_lack, group_lack1, group_lack2, group_lack3)
print(f'등분산성 검정 결과: W={stat:.3f}, p={p:.4f}')

# 정규성 가정과 등분산성 가정을 충족하지 않음. 
```

```python
# 정규성, 등분산성 가정을 위반하고, 순서형 변수이기 때문에 kruskal-wallis 검정을 사용함. 
from scipy.stats import kruskal

stat, p = kruskal(group_no_lack, group_lack1, group_lack2, group_lack3)
print(f"Kruskal-Wallis H-test: H={stat:.3f}, p={p:.4f}")

# 네 그룹 간 overall_satis(전반적 만족도)의 분포 차이가 통계적으로 유의미함을 의미
```

```python
mean_satis = df2324.groupby(['wellness', 'infolack'])['overall_satis'].mean()
print(mean_satis)
```

```python
# 맨-위트니: 웰니스 관광객과 일반관광객 사이에 전반적 만족도 점수가 통계적으로 의미있게 다른가?
from scipy.stats import mannwhitneyu

group_wellness = df2324[df2324['wellness'] == 1]['overall_satis'].dropna()
group_general = df2324[df2324['wellness'] == 0]['overall_satis'].dropna()

stat, p = mannwhitneyu(group_wellness, group_general, alternative='two-sided')
print(f"Mann-Whitney U test: statistic={stat}, p-value={p:.4f}")

# 두 그룹의 만족도는 서로 다르다고 할 수 있음. 
```
