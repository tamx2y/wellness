```python
import pandas as pd

# 엑셀 파일 경로
file_path = '웰니스_리뷰_거제벨버디어.xlsx'

# 엑셀 파일 읽기
df = pd.read_excel(file_path)

# 데이터 미리보기
print(df.head())
```

```python
# 중복 행 제거 (전체 컬럼 기준)
df = df.drop_duplicates()

# 결과 확인
print(f"중복 제거 후 행 개수: {len(df)}")
print(df.head())

# 엑셀로 저장
df.to_excel('웰니스_리뷰_거제벨버디어_중복제거.xlsx', index=False)
```
