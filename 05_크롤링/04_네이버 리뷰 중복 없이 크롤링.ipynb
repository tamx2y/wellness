```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import re

# 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
results = []

# 직접 수집한 네이버 지도 URL 목록
place_urls = [
    
    {'업소명': '센텀스파랜드', '지역': '부산',
     'url': 'https://pcmap.place.naver.com/place/13359851/review/visitor'}
]

def extract_reviews():
    reviews = []
    time.sleep(2)

    review_blocks = driver.find_elements(By.CSS_SELECTOR, 'ul#_review_list > li')
    print(f"리뷰 element 개수: {len(review_blocks)}")

    for item in review_blocks:
        try:
            review_text = item.find_element(By.CSS_SELECTOR, 'div.pui__vn15t2 a').text
            writer = item.find_element(By.CSS_SELECTOR, 'span.pui__NMi-Dp').text

            # 작성자 정보 (ex. 리뷰 28 · 사진 4 · 팔로워 1)
            try:
                writer_info = item.find_element(By.CSS_SELECTOR, 'div.pui__RuLAax').text
            except:
                writer_info = None

            # 날짜
            try:
                date_element = item.find_elements(By.CSS_SELECTOR, 'span.pui__gfuUIT span.pui__blind')
                review_date = date_element[1].text  # 두 번째 요소가 실제 날짜
            except:
                review_date = None

            # 방문횟수 (숫자만 추출)
            try:
                nth_visit_text = item.find_elements(By.CSS_SELECTOR, 'span.pui__gfuUIT')[1].text  # 예: "2번째 방문"
                nth_visit = int(re.search(r'\d+', nth_visit_text).group())  # 숫자만 추출
            except:
                nth_visit = None

            # 인증 수단 (예: '네이버페이')
            try:
                verify_elements = item.find_elements(By.CSS_SELECTOR, 'div.pui__QKE5Pr > span.pui__gfuUIT')
                if len(verify_elements) >= 3:
                    verify_how = verify_elements[2].text.replace('인증 수단', '').strip()
                else:
                    verify_how = None
            except:
                verify_how = None

            reviews.append({
                '리뷰': review_text,
                '작성자': writer,
                '작성자정보': writer_info,
                '작성일': review_date,
                '방문횟수': nth_visit,
                '인증수단': verify_how
            })
        except Exception as e:
            print("리뷰 수집 에러:", e)
            continue

    return reviews


# 1. 페이지 이동
driver.get(place_urls[0]['url'])
time.sleep(2)

click_count = 0
max_clicks = 141

# 2. '다음' 버튼 반복 클릭
while click_count < max_clicks:
    try:
        time.sleep(2)
        next_btn = driver.find_element(By.CSS_SELECTOR, 'a.fvwqf')
        if 'disabled' in next_btn.get_attribute('class'):
            print("마지막 페이지입니다.")
            break
        next_btn.click()
        click_count += 1
        print(f"다음 버튼 클릭! ({click_count}회)")
    except:
        print("더보기 완료")
        break

# ✅ 3. 모든 리뷰가 펼쳐진 뒤, 한 번만 전체 수집
review_data = extract_reviews()
for r in review_data:
    r['업소명'] = place_urls[0]['업소명']
    r['지역'] = place_urls[0]['지역']
results.extend(review_data)

# 2. 저장
df = pd.DataFrame(results)
df.to_excel('웰니스_리뷰_센텀스파랜드.xlsx', index=False)
print("리뷰 저장 완료!")
driver.quit()
```
